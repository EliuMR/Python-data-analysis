```python
import seaborn as sns
%matplotlib inline
tips= sns.load_dataset('tips')
tips.head()
```

```python
sns.set_style('darkgrid')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
```

```python

sns.countplot(x='sex',data=tips)
sns.despine()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

```python
#El método sns.set_context() en Seaborn se utiliza para ajustar la escala y el 
#contexto de los gráficos. Esto es útil cuando necesitas adaptar el tamaño de los 
#elementos gráficos para diferentes tipos de visualización, como presentaciones,
#informes o análisis en pantalla
sns.set_context('poster',font_scale=0.5)
sns.countplot(x='sex',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='plasma')
```
