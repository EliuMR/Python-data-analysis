```python
import seaborn as sns
%matplotlib inline
tips= sns.load_dataset('tips')
tips.head()
```

```python
sns.set_style('darkgrid')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
```

```python

sns.countplot(x='sex',data=tips)
sns.despine()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

```python
#El mtodo sns.set_context() en Seaborn se utiliza para ajustar la escala y el 
#contexto de los grficos. Esto es til cuando necesitas adaptar el tamao de los 
#elementos grficos para diferentes tipos de visualizacin, como presentaciones,
#informes o anlisis en pantalla
sns.set_context('poster',font_scale=0.5)
sns.countplot(x='sex',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='plasma')
```
