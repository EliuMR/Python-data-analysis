```python
import seaborn as sns
%matplotlib inline
iris= sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
#se utiliza para crear una matriz de gráficos de pares que muestra las relaciones
#entre las diferentes combinaciones de variables en un conjunto de datos.
sns.pairplot(iris)
```

```python
#El objeto sns.PairGrid de Seaborn es una herramienta flexible para crear gráficos de
#pares personalizados. A diferencia de sns.pairplot(), que tiene una funcionalidad
#predefinida, sns.PairGrid permite un control detallado sobre cada tipo de gráfico en 
#la matriz de gráficos de pares.
import matplotlib.pyplot as plt
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.displot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
#La clase sns.FacetGrid se utiliza para crear una cuadrícula de gráficos faceteados basada en variables categóricas.
# Crear un objeto FacetGrid
g = sns.FacetGrid(data=tips, col='time', row='smoker')

# Mapear el histograma a la cuadrícula
g.map(sns.histplot, 'total_bill')
```

```python
# Crear un objeto FacetGrid
g = sns.FacetGrid(data=tips, col='time', row='smoker')

# Mapear el histograma a la cuadrícula
g.map(sns.histplot, 'total_bill','tip')
```
