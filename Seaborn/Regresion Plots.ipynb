```python
import seaborn as sns
%matplotlib inline
tips= sns.load_dataset('tips')

tips.head()
```

```python
#El método sns.lmplot() de Seaborn se utiliza para crear gráficos de dispersión (scatter plots) con líneas de regresión ajustadas. 
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'])
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex') #separar por columns 
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,row='day',hue='sex') 
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',aspect=0.6) 
```
