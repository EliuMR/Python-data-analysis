```python
import seaborn as sns
%matplotlib inline
tips= sns.load_dataset('tips')
tips.head()
```

```python
#promedio
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
```

```python
#desviación estándar
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips) # cuenta 
```

```python
#box plot
#Se crea un gráfico de caja que muestra la distribución de la cuenta total (total_bill) para cada día de la semana (day) en el dataset tips.
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python

#Un gráfico de violín (violin plot) es otra herramienta gráfica que permite visualizar 
#la distribución de un conjunto de datos. Similar a un box plot, muestra la distribución
#de los datos a través de sus cuartiles, pero también incluye una estimación de la 
#densidad de probabilidad de los datos a diferentes valores
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
#Un gráfico de puntos (strip plot) es una visualización que muestra la distribución de
#puntos a lo largo de un eje categórico. Es útil para mostrar todos los puntos de datos
#y puede ser particularmente efectivo cuando se desea observar la densidad y la 
#dispersión de los datos.
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
#combinación de stripplot y violinplot
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.factorplot(x='day',y='total_bill',data=tips)
```
