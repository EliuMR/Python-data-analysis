## Plotly and Cufflinks

```python
import pandas as pd
import numpy as np
```

```python
from plotly import __version__
print(__version__)
```

```python
import cufflinks as cf
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
```

```python
init_notebook_mode(connected=True)
```

```python
cf.go_offline()
```

```python
#data
df=pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
```

```python
df.head()
```

```python
df2=pd.DataFrame({'Caregory':['A','B','C'],'Values':[32,43,50]})
df2
```

```python
df.plot()
```

```python
df.iplot() #interactive plot
```

```python
df.iplot(kind='scatter',x='A',y='B',mode='markers',size=20)
```

```python
df2.iplot(kind='bar',x='Caregory',y='Values')
```

```python
df.count().iplot(kind='bar') #graficas usando funcines de agregado
```

```python
df.sum().iplot(kind='bar') 
```

```python
df.iplot(kind='box')
```

```python
#3d superficies
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,40,50],'z':[100,200,300,400,500]})
```

```python
df3
```

```python
df3.iplot(kind='surface')
```

```python
df['A'].iplot(kind='hist',bins=100) #histrograma
```

```python
df.iplot(kind='hist',bins=50) #histrograma de todas las columnas
```

```python
import numpy as np
df.iplot(kind='bubble',x='A',y='B',size='C')
#coordenadas A,B con tamao del crculo a una columna en este caso C
```
