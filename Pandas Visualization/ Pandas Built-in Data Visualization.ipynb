```python
#El método plot() de pandas se utiliza para generar gráficos a partir de un DataFrame o una Serie.
df1['A'].plot(kind='hist')
```

```python

#El método plot.area() de pandas se utiliza para crear un gráfico de área a partir de un 
#DataFrame. Un gráfico de área es similar a un gráfico de líneas, pero las áreas bajo las 
#líneas están rellenas. Este tipo de gráfico es útil para visualizar cómo cambia una o más 
#series de datos a lo largo del tiempo o cualquier otra variable continua.
df2.plot.area(alpha=0.2)
```
