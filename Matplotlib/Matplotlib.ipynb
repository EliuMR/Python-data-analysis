## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
#Functional
plt.plot(x,y,'r.')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gr√°fica')
plt.show()
```

```python
#multi
plt.subplot(1,2,1) # number rows, number columns, plot refer
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# object oriented
fig = plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8]) #: This list specifies the position and size of
#the axes within the figure. The values are in the format [left, bottom, width, 
#height] and are expressed as fractions of the figure's width and height.
axes.plot(x,y)
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title('T')
axes2.set_title('R')
```

```python
fig = plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
fig,axes= plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
#axes.plot(x,y)
```

```python
fig,axes= plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('1')
axes[1].plot(y,x)
axes[1].set_title('2')
```

### Figure size and DPI

```python
fig = plt.figure(figsize=(4,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(figsize=(4,2))
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout() # que se distingan los numeros cuando son tapados por otra gr√°fica
```

```python
fig = plt.figure(figsize=(4,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label='1')
ax.plot(y,x,label='2')
ax.legend(loc=10) # cada n√∫mero representa una ubicaci√≥n en la tabla
```

```python
fig = plt.figure(figsize=(4,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label='1')
ax.plot(y,x,label='2')
ax.legend(loc=(0.1,0.1)) # especificamos persolanizada
```

### Plot Apparance

```python
fig=plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#ff8c00') #color or in hex color
```

```python
fig=plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=5,alpha=0.4) #alpha transparency of the line
                                                  #lw=linewidth
```

```python
fig=plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=2,linestyle='dotted') 
```

```python
fig=plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=2,linestyle='dotted',marker='o',markersize=20,
       markerfacecolor='red',markeredgewidth=2,markeredgecolor='green')
```

```python
fig=plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=2,linestyle='--')
#limits
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
